## Correcting for Venus' Position Relative to the Sun 

```python
Image(filename='kepler-third-law.jpg')
```

$ r = a\frac{1-e^2}{1+ecos\theta}$  

Then you use the approximation $\frac{1}{1+x} \approx 1-x $:  

$r = a(1-e^2)(1 - ecos(\frac{360}{365.256363}day))$

Then we correct for the perihelion being on Jan 2 this year:

$ r = 1 - 0.01672 cos(0.9856(day-2))$

** Finding the distance between Earth and the Sun **
1: mean distance between Earth and the Sun  
0.01672: eccentricty of Earth's orbit around the Sun  
360: number of deg in full rotation  
365.2563: the length of sidereal year, in mean solar days  
360/365.2563 = 0.9856  
day: day number of the year  
Jan 2: is the day of the year that the earth reaches the perihelion  

#### The Law of Cosines
$ cos(A) = \frac{b^2 + c^2 - a^2}{2bc} $  

$ cos(B) = \frac{a^2 + c^2 - b^2}{2ac} $  

$ cos(C) = \frac{a^2 + b^2 - c^2}{2ab}$  

a: Earth and Venus  
b: Earth and the Sun  
c: Venus and the Sun  
A: Sun  
B: Venus  
C: Earth

```python
Image(filename = 'Law_of_Cosines1.jpg')
```

```python
def correction(day, sun_dec, sun_ra, ven_ra, ven_dec, a):
    # finding the distance between Earth and the Sun
    b = 1 - 0.01672*np.cos(0.9856*(day - 2))
    
    # finding the distance between the Sun and Venus
    c1 = SkyCoord(ra=sun_ra*u.degree, dec=sun_dec*u.degree, distance=b*u.au, frame='icrs')
    c2 = SkyCoord(ra=ven_ra*u.degree, dec=ven_dec*u.degree, distance=a*u.au, frame='icrs')
    c = c1.separation_3d(c2).value
    
    # plugging into the law of cosines to find the angle between the Sun and Venus
    distance =  np.arccos((a**2 + c**2 - b**2)/(2*(a)*(c)))
    return distance
```

```python
# Date 1 or min
# Sun's right ascension and declination http://www.stargazing.net/mas/planet4.htm
d1 = correction(187, 1.1254, 7.4318, venus_min_ra, venus_min_dec, venus_min_range)
d1
```

```python
#Date 2 or max 
d2 = correction(260, 6.0059, 23.2612, venus_max_ra, venus_max_dec, venus_max_range)
d2
```

```python
d2_correction = venus_max_dec - d2
d2_correction
```

```python
d1_correction = venus_min_dec - d1
d1_correction
```
